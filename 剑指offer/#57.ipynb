```
class Solution {
public:
    bool binary_search(const vector<int>& a, int l, int r, const int target) {
        while(l <= r) {
            int mid = (l + r) >> 1;
            if(a[mid] == target)
                return true;
            else if(a[mid] < target)
                l = mid + 1;
            else
                r = mid - 1;
        }
        return false;
    }

    vector<int> twoSum(vector<int>& nums, int target) {
        return two_point(nums, target);
        // return twoSum_v2(nums, target);
        vector<int> res;
        if(nums.empty())
            return res;
        for(int i = 0; i < nums.size(); ++i) {
            if(binary_search(nums, 0, i - 1, target - nums[i]) || 
                binary_search(nums, i + 1, nums.size() - 1, target - nums[i])) {
                    res.push_back(nums[i]);
                    res.push_back(target - nums[i]);
                    break;
                }
        }
        return res;
    }
    vector<int> two_point(vector<int>& a, int target) {
        vector<int> res;
        if(a.empty())
            return res;
        int l = 0, r = a.size() - 1;
        while(l < r) {
            int val = a[l] + a[r];
            if(val == target) {
                res.push_back(a[l]);
                res.push_back(a[r]);
                break;
            } else {
                if(val > target) // 为什么不用l-- 因为 l - 1, r 已经判断过了 不符合才会到当前这一步
                    r--;
                else // 同理 l, r + 1 一定已经判定过 
                    l++;
            }
        }
        return res;
    }

    vector<int> twoSum_v2(vector<int>& nums, int target) {
        vector<int> res;
        if(nums.empty())
            return res;
        unordered_map<int, int> m;
        for(int i = 0; i < nums.size(); i++) {
            auto ptr = m.find(nums[i]);
            if(ptr == m.end())
                m[nums[i]] = 1;
            else
                ptr->second++;
        }
        for(int i = 0; i < nums.size(); i++) {
            if(m.find(target - nums[i]) != m.end()) {
                if(target == (nums[i] << 1)) {
                    if(m[nums[i]] >= 2) {
                        res.push_back(nums[i]);
                        res.push_back(nums[i]);
                        break;
                    }
                } else {
                    res.push_back(nums[i]);
                    res.push_back(target - nums[i]);
                    break;
                }
            }
        }
        return res;
    }
};
```
